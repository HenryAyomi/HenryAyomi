```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from time import sleep
from random import randint
info=[]
for i in range(0,21):
    url= f"https://www.propertypro.ng/property-for-sale/house/in/lagos?search=&auto=&mode_hidden=&typed_hidden=&bedroom=&min_price=3000000&max_price=100000000&page={i}"
    
    response=requests.get(url)
    

    response=response.content
    soup=BeautifulSoup(response,"html.parser")
    div=soup.find("div",itemtype="https://schema.org/ItemList")
    housing=div.findAll("div",class_="single-room-sale listings-property")
    
    for house in housing:
        name=house.find(["h3","h4"],class_="listings-property-title2").text if house.find("h3",class_="listings-property-title2") else "Not FOUND"
        price=house.find("h3",class_="listings-price").text if house.find("h3",class_="listings-price") else np.nan
        location=house.find_all("h4")
        try:
            location=location[1].text
        except:
            pass
            
        #find_room=house.find_all("span") if house.find("span",[]) else[]
        #room=find_room
        find_=house.find("div",class_="fur-areea")
        try:
            room=(list(find_))[1].text
        except:
            pass
        try:
            bathroom=(list(find_))[3].text
        except:
            pass
        try:
            toilet=(list(find_))[5].text
        except:
            pass
        info.append([name, price, location, room, bathroom, toilet])
df=pd.DataFrame(info,columns=["Name","Price","Location","Bedroom","Bathroom","Toilet"])   
df
        
            
```
